```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Reading image 
img = cv2.imread('/content/drive/MyDrive/Coding Assignments IVP/colorful.jpeg')    
# img is a 3-dimensional numpy array (third number indicates channel)
print(type(img)) 
print(img.shape)  
```

```python
# In order to display image in RGB format we need to convert it from BGR to RGB
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Seperating different channels 
# for RED channel
img_red = img_rgb[:,:,0]
# for GREEN channel
img_green = img_rgb[:,:,1]
# for BLUE channel
img_blue = img_rgb[:,:,2]

plt.figure(figsize = (15,15))
plt.subplot(1,3,1)
plt.imshow(img_red, cmap = 'gray')    
plt.title('R')
plt.subplot(1,3,2)
plt.imshow(img_green, cmap = 'gray')
plt.title('G')
plt.subplot(1,3,3)
plt.imshow(img_blue, cmap = 'gray')
plt.title('B')
```

As we can see, each displayed gray image represents the intensity values of a single color channel. The intensity values in each channel determine the color and brightness of corresponding pixel in RGB image. <br>
The gray scale values of the individual channel images give information about the dominance of a specific color in different part of the image. 

```python
# Convering RGB image to HSV format
img_hsv = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
# Writing the result
cv2.imwrite('/content/drive/MyDrive/Coding Assignments IVP/colorful_hsv.jpg',img_hsv)
```

```python
# Seperating each channel
# for RED channel
img_h = img_hsv[:,:,0]
# for GREEN channel
img_s = img_hsv[:,:,1]
# for BLUE channel
img_v = img_hsv[:,:,2]

plt.figure(figsize = (15,15))
plt.subplot(1,3,1)
plt.imshow(img_h, cmap = 'gray')    
plt.title('H')
plt.subplot(1,3,2)
plt.imshow(img_s, cmap = 'gray')
plt.title('S')
plt.subplot(1,3,3)
plt.imshow(img_v, cmap = 'gray')
plt.title('V')
```

As we can see, each displayed gray image, represent different aspect. <br>
Hue: shows the color information of the image. This channel identifies the dominant color in the image. 
Saturation: represents the intensity of the color in the image, where higher saturation values indicate more vivid and pure colors, while lower saturation values indicate less vivid colors. 
Value: represents the brightness of the image, where higher values indicate brighter pixels and lower values indicate darker pixels.  

```python
# The range of blue pixels in hue is [110,130]
# We seperate them to lower range and upper range
l1_blue = np.array([110,0,0])
u1_blue = np.array([130, 255, 255])

mask = cv2.inRange(img_hsv, l1_blue, u1_blue)
result = cv2.bitwise_and(img_rgb,img_rgb, mask=mask)
plt.imshow(result)
```
